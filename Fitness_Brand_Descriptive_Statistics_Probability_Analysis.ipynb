```python
# Contingency Table for Product and Gender
pd.crosstab(index=data["Product"], columns=data["Gender"], margins=True, )
```

```python
#Conditional Probability of Products given gender P(Product | Gender)
pd.crosstab(data.Product, data.Gender, normalize='columns')
```

We can infer from the above contingency table following things,


*   Probability of buying KP281 given a female customer is 52.63%.
*   Probability of buying KP281 given a male customer is 38.46%.



```python
#Conditional Probability of Gender given Product P(Gender | Product)
pd.crosstab(data.Product, data.Gender, normalize='index')
```

We can infer from the above contingency table following things,


*   Probability of a female buying given that it is KP781 product is 17.5%.
*   Probability of a male buying given that it is KP781 product is 82.5%.

```python
#checking range of values for Income
data.sort_values(by = ['Income'], ascending=True)
```

```python
#conversion of Income to categorical attributes to analyze
dur_point=[1,20000,40000,60000,80000,100000,120000]
dur_labels=["under 20K","20K-40K","40K-60K","60K-80K","80K-100K","100K above"]
data["Income_Slab"]=pd.cut(data["Income"],
bins=dur_point,
labels=dur_labels) 
data       
```

```python
#Product wise buyers income range
plt.figure(figsize=(8,6))
sns.countplot(x='Product', hue='Income_Slab', data=data)
plt.ylabel("Frequency of order")
plt.title("Product wise buyers income range distribution")
plt.show()
```

From the above chart we can infer that for KP281 & KP481, the highest portion of customer has income range is 40K-60 K and for KP781 it is 80K-100K

```python
#checking range of values for Miles
data.sort_values(by = ['Miles'], ascending=True)
```

```python
#conversion of expected Miles to run per week to categorical attributes to analyze
dur_point=[1,50,100,150,200,250,300,360]
dur_labels=["1-50 Miles","51-100 Miles","101-150 Miles","151-200 Miles","201-250 Miles","251-300 Miles", "300-360 Miles"]
data["Miles_Slab"]=pd.cut(data["Miles"],
bins=dur_point,
labels=dur_labels) 
data       
```

```python
#Product wise buyers expected Miles to run per week range
plt.figure(figsize=(8,6))
sns.countplot(x='Product', hue='Miles_Slab', data=data)
plt.ylabel("Frequency of order")
plt.title("Product wise buyers expected Miles to run per week range")
plt.show()
```

From the above chart we can infer that for KP281 & KP481, the highest portion of customer has expected Miles run per week is 51-100 Miles and for KP781 it is 151-200 Miles.

```python
#For correlation of Product with other atributes Education, Usage, Fitness, Income and Mile
sns.pairplot(data=data,hue="Product")
plt.title("correlation of Product with other atributes Education, Usage, Fitness, Income and Mile")
plt.show()
```

From the above pair plot we can infer that,


*   KP781 is preffered among 25 to 35 years age group.
*   Customers with higher education going for KP281 product.
*   Medium range usage (3-5) going for KP481 product.
